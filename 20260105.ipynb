```python
# 성적 처리 프로그램

score = int(input("성적을 입력해주세요: "))

grade = ''

if score >= 90 and score <=100:
    grade = 'A'
elif score < 90 and score >= 80: 
    grade = 'B'
elif score < 70 and score >= 60: 
    grade = 'C'
elif score < 60 and score >= 50: 
    grade = 'D'
elif score > 100 : 
    grade = '오류입니다'
else: 
    grade = 'F'

print(grade)
```

```python
#업앤 다운 게임 만들기

import random
answer = random.randint(1, 100)

count = 0 

while count <= 10 :

    guess = int(input("기회는 10번, 숫자를 추측해보세요"))
    count += 1

    if guess < answer:
        print(f"답은 {guess}보다 더 큰 숫자입니다")
    elif guess > answer:
        print(f"답은 {guess}보다 작은 숫자입니다")
    else: 
        print("정답입니다.")
        break
else:  
    print("정답을 10번 이내로 맞추지 못하셨습니다")
```

```python
# 다른 미션
# n을 입력 받은 뒤
# 트리 높이는 최대 10층, 
# 별을 출력하되 별의 갯수만큼 출력하다가 별이 다 소진되면 멈추는 코드작성

# floor = 1
# result = int(input("별을 얼마나 달까요? "))

n = int(input("얼마나 별을 달까요"))
star_count = 0


for i in range(1,n):
    star_remain = n - star_count
    if star_remain > i: 
        print('*'*i)
        star_count += i
    else: 
        print('*'*star_remain)
        break

```
